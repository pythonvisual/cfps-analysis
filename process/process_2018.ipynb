```python
import pandas as pd

df = pd.read_csv('data/2018/18relation.csv',dtype={'fullVisitorId': 'str'})
#显示所有行，把列显示设置成最大
# pd.set_option('display.max_rows', 30)
# pd.set_option('display.max_rows', None)
df.describe().T 
```

```python
col_n = ['pid','fid18','hukou_a18_s','tb4_a18_s','tb1y_a_s','tb1m_a_s','familysize18','pid_a_m','pid_a_f','pid_a_s']
df = pd.DataFrame(df,columns = col_n)
df.describe().T

```

```python
df.rename(columns={"fid18":"fid","hukou_a18_s":'mate_hukou','tb4_a18_s':'mate_edu','familysize18':'familysize','pid_a_f':'pid_f','pid_a_m':'pid_m','pid_a_s':'mate_pid'},inplace=True)
df.describe().T
def getMageAge(birthYear):
    if birthYear <= 0:
        return birthYear
    else:
        return (2014-birthYear)
df['mate_age'] = df.apply(lambda x: getMageAge(x.tb1y_a_s), axis=1)
```

```python
col_n = ['pid','fid','mate_hukou','mate_edu','mate_age','familysize','pid_f','pid_m','mate_pid']
df = pd.DataFrame(df,columns = col_n)
df.describe().T
```

```python
## 处理个人数据
personDf = pd.read_csv('data/2018/2018.csv')
col_n = ['pid','qg12','provcd18','edu_last','fid18','age','urban18','qa301','qea0','qp201','gender','qi101','qi200','qi2001','qi1011','fid14','qi200','qi2001','qi201_s_1','qi201_s_2','qi201_s_3','qi201_s_4','qi201_s_5','qi301_s_1','qi301_s_2','qi301_s_3','qi301_s_4','qi301_s_5','qi301_s_6','qi301_s_7','qi301_a_1','qi301_a_2','qi301_a_3','qi301_a_4','qi301_a_7','retire','pension']
personDf = pd.DataFrame(personDf,columns = col_n)
personDf.rename(columns={'fid18':'fid','qg12':'salary','provcd18':'province','edu_last':"edu","qa301":"hukou","qea0":"marry","qp201":"health","urban18":"urban","cfps_gender":"gender"},inplace=True)

def isRetire(x):
    if x.retire == 1 :
        return 1
    elif x.pension == 1 :
        return 1     
    # 如果都小于0，才能是为无效数据
    elif (x.retire < 0) & (x.pension < 0) :
        return -8
    else:
        return 0
personDf['is_retire'] = personDf.apply(lambda x:isRetire(x), axis=1)
personDf.describe().T

```

```python
# newDf = pd.merge(df,personDf,on = ['pid','fid'],how="outer")
newDf = pd.merge(df,personDf,on = ['pid','fid'])
newDf.describe().T
# newDf.groupby(["fid"]).count()
```
