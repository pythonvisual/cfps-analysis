```python
import pandas as pd

df = pd.read_csv('data/2016/16relation.csv')
pd.set_option('display.max_columns', 50); 
pd.set_option('display.width', 1000) #将结果完整输出
#显示所有行，把列显示设置成最大
# pd.set_option('display.max_rows', None)
# df.describe().T 
```

```python
col_n = ['pid','fid16','tb1y_a_s','tb1m_a_s','tb1y_a_m','code_a_f','pid_f','tb1y_a_f','code_a_m','pid_m','tb4_a16_m','tb4_a16_f','tb4_a16_s','hukou_a16_s','hukou_a16_f','hukou_a16_m','pid_s']
df = pd.DataFrame(df,columns = col_n)
df.describe().T
```

```python
df.rename(columns={"fid16":"fid","hukou_a16_s":'mate_hukou','tb4_a16_s':'mate_edu','hukou_a16_f':'hukou_parent_f','hukou_a16_m':'hukou_parent_m','tb4_a16_f':'parent_f_edu','tb4_a16_m':'parent_m_edu','pid_s':'mate_pid'},inplace=True)
df.describe().T
def getMageAge(birthYear):
    if birthYear <= 0:
        return birthYear
    else:
        return (2016-birthYear)
df['mate_age'] = df.apply(lambda x: getMageAge(x.tb1y_a_s), axis=1)
df['parent_f_age'] = df.apply(lambda x: getMageAge(x.tb1y_a_f), axis=1)
df['parent_m_age'] = df.apply(lambda x: getMageAge(x.tb1y_a_m), axis=1)
# print(df)

col_n = ['pid','fid','pid_f','hukou_parent_f','parent_f_age','parent_f_edu','pid_m','hukou_parent_m','parent_m_edu','parent_m_age','mate_edu','mate_hukou','mate_age','mate_pid']
df = pd.DataFrame(df,columns = col_n)
df.describe().T
```

```python
import pandas as pd

personDf = pd.read_csv('data/2016/2016.csv')

col_n = ['pid','qg12','provcd16','cfps2016edu','cfps_age','urban16','pa301','qea0','qp201','cfps_gender','qi101','qi200','qi2001','qi1011','fid16','qi201_s_1','qi201_s_2','qi201_s_3','qi201_s_4','qi201_s_5','qi301_s_1','qi301_s_2','qi301_s_3','qi301_s_4','qi301_s_5']
personDf = pd.DataFrame(personDf,columns = col_n)

def isRetire(x):
    if x.qi101 == 1 :
        return 1
    elif x.qi200 == 1 :
        return 1     
    elif x.qi2001 == 1 :
        return 1   
    elif x.qi1011 == 1 :
        return 1             
    # 如果都小于0，才能是为无效数据
    elif (x.qi101 < 0) & (x.qi200 < 0) & (x.qi2001 < 0) & (x.qi1011 < 0) :
        return -8
    else:
        return 0
personDf['is_retire'] = personDf.apply(lambda x:isRetire(x), axis=1)

personDf.rename(columns={'fid16':'fid','pa301':'hukou','qg12':'salary','provcd16':'province','cfps2016edu':"edu","cfps_age":"age","qea0":"marry","qp201":"health","urban16":"urban","cfps_gender":"gender"},inplace=True)
personDf.describe().T

col_n = ['pid','salary','province','edu','age','urban','hukou','marry','health','gender','fid','qi301_s_1','qi301_s_2','qi301_s_3','qi301_s_4','qi301_s_5','is_retire']
personDf = pd.DataFrame(personDf,columns = col_n)
personDf.describe().T
```

```python
newDf = pd.merge(df,personDf,on = ['pid','fid'])
newDf.describe().T
```
