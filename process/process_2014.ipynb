```python
# https://zhuanlan.zhihu.com/p/280892036
# https://zhuanlan.zhihu.com/p/499600704
import pandas as pd
import statsmodels.formula.api as smf #方法二
import numpy

df = pd.read_csv('data/2014/2014familyRelation.csv')
#显示所有行，把列显示设置成最大
# pd.set_option('display.max_rows', None)
df.describe().T 
```

```python
col_n = ['pid','pid_f','pid_m','fid14','qa301_a14_s','tb4_a14_s','tb1y_a_s','tb1m_a_s','cyear','cmonth','pid_s']
df = pd.DataFrame(df,columns = col_n)

df.describe().T

```

```python
df.rename(columns={"fid14":"fid","qa301_a14_s":'mate_hukou','tb4_a14_s':'mate_edu','pid_s':'mate_pid'},inplace=True)
df.describe().T
def getMageAge(birthYear):
    if birthYear <= 0:
        return birthYear
    else:
        return (2014-birthYear)
df['mate_age'] = df.apply(lambda x: getMageAge(x.tb1y_a_s), axis=1)
```

```python
col_n = ['pid','fid','mate_hukou','mate_edu','mate_age','pid_f','pid_m','mate_pid']
df = pd.DataFrame(df,columns = col_n)
df.describe().T
```

```python
import pandas as pd
personDf = pd.read_csv('data/2014/2014.csv')

col_n = ['pid','qg12','provcd14','cfps2012_latest_edu','cfps2014_age','urban14','qa301','qea0','qp201','cfps_gender','qi101','qi200','qi2001','qi1011','fid14','qi201_s_1','qi201_s_2','qi201_s_3','qi201_s_4','qi201_s_5','qi301_s_1','qi301_s_2','qi301_s_3','qi301_s_4','qi301_s_5','qi301_s_6','qi301_s_7','retirement2012']
personDf = pd.DataFrame(personDf,columns = col_n)

def isRetire(x):
    if x.qi101 == 1 :
        return 1
    elif x.retirement2012 == 1 :
        return 1           
    # 如果都小于0，才能是为无效数据
    elif (x.qi101 < 0) & (x.retirement2012 < 0):
        return -8
    else:
        return 0
personDf['is_retire'] = personDf.apply(lambda x:isRetire(x), axis=1)
```

```python
personDf.rename(columns={'fid14':'fid','provcd14':'province','qg12':'salary','cfps2012_latest_edu':"edu","cfps2014_age":"age","qa301":"hukou","qea0":"marry","qp201":"health","urban14":"urban","cfps_gender":"gender"},inplace=True)
personDf.describe().T
```

```python
newDf = pd.merge(df,personDf,on = ['pid','fid'])
newDf.describe().T
```
